```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fp = '/content/drive/Shareddrives/La(LA)LAB/Research Projects/kc-coding/Data/Cleaned_Mel_CPS_19_Dataset.csv'
data = pd.read_csv(fp)
```

```python
data
```

```python
# prompt: visualize and plot the histogram of data

# Assuming 'data' is your DataFrame and you want a histogram of a specific column, replace 'column_name'
plt.figure(figsize=(10, 6))
sns.histplot(data['KC'], kde=True)  # kde=True adds a kernel density estimate
plt.title('Histogram of Column Name')
plt.xlabel('Column Name')
plt.ylabel('Frequency')
plt.show()
```

```python
# prompt: plot normalized frequency 'KC' column grouped by 'task' in data

# Normalize the frequency of 'KC' within each 'task' group
normalized_kc = data.groupby('task')['KC'].value_counts(normalize=True).reset_index(name='normalized_frequency')

# Create the plot
plt.figure(figsize=(12, 6))
sns.barplot(x='task', y='normalized_frequency', hue='KC', data=normalized_kc)
plt.title('Normalized Frequency of KC by Task')
plt.xlabel('Task')
plt.ylabel('Normalized Frequency')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for readability
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
```
